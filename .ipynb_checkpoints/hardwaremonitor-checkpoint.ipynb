```python
import pandas as pd
df = pd.read_csv('~/Desktop/HardwareMonitoring.csv', sep=",")
df.head()
```

```python
df.columns = ['q','date','gpu_temp','gpu_voltage','gpu_fan','gpu_usage','mem_usage']
df['gpu_voltage'] = pd.to_numeric(df['gpu_voltage'], errors='coerce')
df['gpu_temp'] = pd.to_numeric(df['gpu_temp'], errors='coerce')
df['gpu_fan'] = pd.to_numeric(df['gpu_fan'], errors='coerce')
df['gpu_usage'] = pd.to_numeric(df['gpu_usage'], errors='coerce')
df['mem_usage'] = pd.to_numeric(df['mem_usage'], errors='coerce')
```

```python
data = df.tail(5000)
data.head()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,3)
# gca stands for 'get current axis'
ax = plt.gca()
#data.plot(kind='line',x='date',y='gpu_temp',ax=ax)
#data.plot(kind='line',x='date',y='gpu_voltage', color='green', ax=ax)
#data.plot(kind='line',x='date',y='gpu_fan', color='orange', ax=ax)
#data.plot(kind='line',x='date',y='gpu_usage', color='purple', ax=ax)
data.plot(kind='line',x='date',y='mem_usage', color='red', ax=ax)

plt.show()
```

```python
plt.rcParams["figure.figsize"] = (20,3)
# gca stands for 'get current axis'
ax = plt.gca()
data.plot(kind='line',x='date',y='gpu_temp',ax=ax)
data.plot(kind='line',x='date',y='gpu_fan', color='orange', ax=ax)
data.plot(kind='line',x='date',y='gpu_usage', color='purple', ax=ax)

plt.show()
```

```python
plt.rcParams["figure.figsize"] = (20,3)
# gca stands for 'get current axis'
ax = plt.gca()
data.plot(kind='line',x='date',y='gpu_voltage', color='green', ax=ax)
plt.show()
```
